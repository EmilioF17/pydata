# 1.usa.gov data from bit.ly

```python
import json;

path = 'ch02/usagov_bitly_data2012-03-16-1331923249.txt';
records = [json.loads(line) for line in open(path)];
```

```python
time_zones = [rec['tz'] for rec in records if rec.has_key('tz')];
```

```python
from collections import defaultdict;

def get_count(seq):
    counts = defaultdict(int);
    for x in seq:
            counts[x] += 1;
    return counts;

counts = get_count(time_zones);

print counts['America/New_York'];
```

```python
def top_counts(count_dict,n=10):
        value_key_pairs = [(counts,tz) for tz,counts in count_dict.items()];
        value_key_pairs.sort();
        return value_key_pairs[-n:];

print top_counts(counts);
```

```python
from collections import Counter;

counts = Counter(time_zones);

print counts.most_common(10);
```

# MovieLens 1M Data Set

```python
import pandas as pd;

def read_dataset(file_name,header_name):
    return pd.read_table(file_name,sep='::',header=None,names=header_name);

file_path = 'ch02/movielens/%s';

unames = ['user_id','gender','age','occuupation','zip'];
rnames = ['user_id','movie_id','rating','timestamp'];
mnames = ['movie_id','title','genres'];

users = read_dataset(file_path % 'users.dat',unames);
ratings = read_dataset(file_path % 'ratings.dat',rnames);
movies = read_dataset(file_path % 'movies.dat',mnames);

#print users[:5];
#print ratings[:5];
#print movies[:5];
```

```python
data = pd.merge(pd.merge(ratings,users),movies);
```

```python
mean_ratings = data.pivot_table('rating',rows='title',cols='gender',aggfunc='mean');
print mean_ratings[:5];
```

```python
ratings_by_title = data.groupby('title').size();
active_titles = ratings_by_title.index[ratings_by_title >=250];
print active_titles;
```

```python
mean_ratings = mean_ratings.ix[active_titles];
print mean_ratings;
```

```python
top_female_ratings = mean_ratings.sort_index(by='F',ascending = False);
print top_female_ratings[:10];
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F'];
sorted_by_diff = mean_ratings.sort_index(by='diff');
print sorted_by_diff[:15];
```

```python
print sorted_by_diff[::-1][:15];
```

```python
rating_std_by_title = data.groupby('title')['rating'].std();
rating_std_by_title = rating_std_by_title.ix[active_titles];
print rating_std_by_title.order(ascending=False)[:10];
```

# US Baby Names 1880-2010

```python
import pandas as pd;

pieces = [];
cols = ['name','sex','births'];

# Load and append into list.
for year in range(1880,2011):
    file_path = 'ch02/names/yob%d.txt' % year;
    frame = pd.read_csv(file_path,names=cols);
    
    frame['year'] = year;
    pieces.append(frame);

names = pd.concat(pieces,ignore_index=True);
print names[:10];
```

```python
total_births = names.pivot_table('births',rows='year',cols='sex',aggfunc='sum');
print total_births[:10];
total_births.plot();
```

```python
def add_prop(group):
        births = group.births.astype(float);
        
        group['prop'] = births/births.sum();
        return group;

names = names.groupby(['year','sex']).apply(add_prop);
print names[:10];
```

```python
def get_top1000(group):
    return group.sort_index(by='births',ascending=False)[:1000];

grouped = names.groupby(['year','sex']);
top1000 = grouped.apply(get_top1000);
```

```python
total_births = top1000.pivot_table('births',rows='year',cols='name',aggfunc=sum);
subset = total_births[['John','Harry','Mary','Marilyn']];

subset.plot(subplots=True,figsize=(12,10),grid=False);
```

```python
table = top1000.pivot_table('prop',rows = 'year', cols = 'sex', aggfunc=sum);
table.plot(yticks = np.linspace(0,1.2,13),xticks=range(1880,2020,10),title="Sum of table1000.prop by year and sex");
```

```python
def get_quantile_count(group, q=0.5):
    prop_cumsum = group.sort_index(by='prop',ascending=False).prop.cumsum();
    return prop_cumsum.searchsorted(q)+1;

diversity = qrouped = top1000.groupby(['year','sex']).apply(get_quantile_count);
diversity = diversity.unstack('sex');
#print diversity.head();
diversity.plot(title="Number of popular names in top 50%");
```
